```python
def zscore(dataframe,colname,ddof=0):
    return (dataframe[colname]-dataframe[colname].mean())/dataframe[colname].std(ddof=ddof)
```

```python
dataset.drop(dataset[zscore(dataset,'Administration') > 3].index,inplace=True)
dataset.drop(dataset[zscore(dataset,'Administration') < -3].index,inplace=True)
```

```python
#Get kurtosis value
dataset.kurtosis()
```

```python
plt.scatter(dataset['Profit'],dataset['Administration'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
X=dataset.values[:,:3]
Y=dataset.values[:,-1]
```

```python
train_x,test_x,train_y,test_y=train_test_split(X,Y,test_size=0.28)
test_y=test_y.reshape(test_y.shape[0],1)
```

```python
#Linear Regression Model
model=LinearRegression().fit(train_x,train_y)
```

```python
model.coef_,model.intercept_
```

```python
#Test the dataset
predictions=model.predict(test_x)
```

```python
#R2 score
r2_score(test_y,predictions,multioutput='variance_weighted')
```
